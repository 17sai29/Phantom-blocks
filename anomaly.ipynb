```python
import os
import csv
import datetime
from tqdm import tqdm

folder_path = r"D:\sai"
output_folder = os.path.join(folder_path, "day_wise_split")
os.makedirs(output_folder, exist_ok=True)

csv_files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith(".csv")]

open_files = {}

for file_path in csv_files:
    print(f"\nProcessing {os.path.basename(file_path)}...")
    
    with open(file_path, 'r', newline='', encoding='utf-8') as f:
        total_lines = sum(1 for _ in f) - 1 

    with open(file_path, 'r', newline='', encoding='utf-8') as infile:
        reader = csv.DictReader(infile)
        
        for row in tqdm(reader, total=total_lines, desc=f"{os.path.basename(file_path)}"):
            try:
                timestamp = int(row['timestamp'])
                dt = datetime.datetime.utcfromtimestamp(timestamp)
                date_str = dt.date().isoformat()  
                
                if date_str not in open_files:
                    output_file_path = os.path.join(output_folder, f"{date_str}.csv")
                    
                    if not os.path.exists(output_file_path):
                        outfile = open(output_file_path, 'w', newline='', encoding='utf-8')
                        writer = csv.DictWriter(outfile, fieldnames=reader.fieldnames)
                        writer.writeheader()
                        open_files[date_str] = (outfile, writer)
                    else:
                        outfile = open(output_file_path, 'a', newline='', encoding='utf-8')
                        writer = csv.DictWriter(outfile, fieldnames=reader.fieldnames)
                        open_files[date_str] = (outfile, writer)
                
                
                outfile, writer = open_files[date_str]
                writer.writerow(row)
            
            except Exception as e:
                print(f"Error processing row: {e}")

for outfile, writer in open_files.values():
    outfile.close()

print(f"\n✅ Day-wise CSVs saved in: {output_folder}")
```

```python
import csv
import os

input_folder = r"D:\sai\day_wise_split_txt"
output_folder = r"D:\sai\address_mapped_final"
mapping_file_path = r"D:\sai\address_mappings"

address_to_id = {}
next_id = 1

if os.path.exists(mapping_file_path):
    with open(mapping_file_path, 'r', encoding='utf-8') as mapping_file:
        for line in mapping_file:
            address, id_ = line.strip().split(',')
            address_to_id[address] = int(id_)
        next_id = max(address_to_id.values(), default=0) + 1

os.makedirs(output_folder, exist_ok=True)

mapping_file = open(mapping_file_path, 'a', encoding='utf-8')

for filename in os.listdir(input_folder):
    if not filename.endswith(".txt"):
        continue

    file_path = os.path.join(input_folder, filename)
    output_file_path = os.path.join(output_folder, filename)

    with open(file_path, 'r', encoding='utf-8') as infile, \
         open(output_file_path, 'w', newline='', encoding='utf-8') as outfile:

        csv_reader = csv.reader(infile, delimiter='\t')
        csv_writer = csv.writer(outfile, delimiter='\t')

        header = next(csv_reader)
        csv_writer.writerow(header)

        from_address_column_index = 3
        to_address_column_index = 4

        for row in csv_reader:
            from_address = row[from_address_column_index]
            to_address = row[to_address_column_index]

            if from_address not in address_to_id:
                address_to_id[from_address] = next_id
                mapping_file.write(f"{from_address},{next_id}\n")
                next_id += 1

            if to_address not in address_to_id:
                address_to_id[to_address] = next_id
                mapping_file.write(f"{to_address},{next_id}\n")
                next_id += 1

            row[from_address_column_index] = address_to_id[from_address]
            row[to_address_column_index] = address_to_id[to_address]

            csv_writer.writerow(row)

    print(f" Mapped and saved: {output_file_path}")

mapping_file.close()

print(f"\n Mapping file fully updated: {mapping_file_path}")
```

```python
#iserror and callinf function mappings
import csv
import os

input_folder = r"D:\sai\address_mapped_final"
output_folder = r"D:\sai\address_mapped_final_2"
function_mapping_file = r"D:\sai\callingfunction_mappings.txt"

calling_function_to_id = {}
if os.path.exists(function_mapping_file):
    with open(function_mapping_file, 'r', encoding='utf-8') as f:
        calling_function_to_id = {func: int(id_) for func, id_ in 
                                  (line.strip().split(',') for line in f if ',' in line)}
next_function_id = max(calling_function_to_id.values(), default=0) + 1

os.makedirs(output_folder, exist_ok=True)

new_function_mappings = []

total_reverted = 0
total_out_of_gas = 0
total_normal = 0

for filename in os.listdir(input_folder):
    if not filename.endswith(".txt"):
        continue

    input_path = os.path.join(input_folder, filename)
    output_path = os.path.join(output_folder, filename)

    with open(input_path, 'r', encoding='utf-8') as infile, \
         open(output_path, 'w', newline='', encoding='utf-8') as outfile:

        reader = csv.reader(infile, delimiter='\t')
        writer = csv.writer(outfile, delimiter='\t')

        header = next(reader)
        writer.writerow(header)

        for row in reader:
            if len(row) <= 13: 
                continue
            calling_function = row[12]
            is_error = row[13].strip().lower()

            func_id = calling_function_to_id.get(calling_function)
            if func_id is None:
                func_id = next_function_id
                calling_function_to_id[calling_function] = func_id
                new_function_mappings.append(f"{calling_function},{func_id}\n")
                next_function_id += 1
            row[12] = func_id

            if is_error == "":
                row[13] = "0"
                total_normal += 1
            elif is_error == "reverted":
                row[13] = "1"
                total_reverted += 1
            elif is_error == "out of gas":
                row[13] = "2"
                total_out_of_gas += 1
            else:
                row[13] = "0"
                total_normal += 1

            writer.writerow(row)

    print(f"Processed and saved: {output_path}")

if new_function_mappings:
    with open(function_mapping_file, 'a', encoding='utf-8') as f:
        f.writelines(new_function_mappings)


print(f"\nFunction mapping file updated: {function_mapping_file}")
print(f"\nTransaction Summary Across All Files:")
print(f"  - Normal transactions (no error): {total_normal:,}")
print(f"  - Reverted transactions: {total_reverted:,}")
print(f"  - Out of gas transactions: {total_out_of_gas:,}")
```

```python
import os
import csv
from collections import Counter

input_folder = r"D:\sai\address_mapped_final_2" 
output_file = r"D:\sai\new_nodes_no_use_with_block.txt"  

from_counter = Counter()
to_counter = Counter()

to_time_block_map = {}

print("First pass: Counting nodes across all files...")

for filename in os.listdir(input_folder):
    if filename.endswith(".txt"):
        filepath = os.path.join(input_folder, filename)
        with open(filepath, 'r', newline='', encoding='utf-8') as f:
            reader = csv.reader(f, delimiter='\t') 
            header = next(reader) 

            block_idx = header.index("blockNumber")
            time_idx = header.index("timestamp")
            from_idx = header.index("from")
            to_idx = header.index("to")

            for row in reader:
                if len(row) <= max(from_idx, to_idx, time_idx):
                    continue  
                from_node = row[from_idx]
                to_node = row[to_idx]
                time_of_tx = row[time_idx]
                block_number = row[block_idx]

                from_counter[from_node] += 1
                to_counter[to_node] += 1

                if to_node not in to_time_block_map:
                    to_time_block_map[to_node] = (time_of_tx, block_number)

print("Second pass: Finding special nodes...")

special_nodes_with_time_block = []

for node, count in to_counter.items():
    if count == 1 and from_counter[node] == 0:
        special_nodes_with_time_block.append((node, to_time_block_map[node][0], to_time_block_map[node][1]))

print(f"Found {len(special_nodes_with_time_block)} special nodes.")

with open(output_file, 'w', encoding='utf-8') as f:
    f.write("node\ttime\tblockNumber\n") 
    for node, time, block in special_nodes_with_time_block:
        f.write(f"{node}\t{time}\t{block}\n")

print(f"Special nodes with block number saved to: {output_file}")
```

```python
import os
import csv
from collections import defaultdict
from tqdm import tqdm

input_folder = r"D:\sai\address_mapped_final_2"
output_file = r"D:\sai\new_nodes_with_creation_gas.txt"

from_nodes = set()
to_counter = defaultdict(int)
to_info_map = {}

all_files = [f for f in os.listdir(input_folder) if f.endswith(".txt")]

print("First pass: Gathering counts and creation data...")

for filename in tqdm(all_files, desc="Reading files"):
    filepath = os.path.join(input_folder, filename)
    with open(filepath, 'r', newline='', encoding='utf-8') as f:
        reader = csv.reader(f, delimiter='\t')
        header = next(reader)

        idx = {
            'block': header.index("blockNumber"),
            'time': header.index("timestamp"),
            'from': header.index("from"),
            'to': header.index("to"),
            'gasUsed': header.index("gasUsed"),
            'gasLimit': header.index("gasLimit")
        }

        for row in reader:
            if len(row) <= max(idx.values()):
                continue  

            from_node = row[idx['from']]
            to_node = row[idx['to']]
            from_nodes.add(from_node)

            to_counter[to_node] += 1

            if to_node not in to_info_map:
                to_info_map[to_node] = (
                    row[idx['time']],
                    row[idx['block']],
                    from_node,
                    row[idx['gasUsed']],
                    row[idx['gasLimit']]
                )

print("Filtering special nodes...")
special_nodes = [
    (node, *to_info_map[node])
    for node, count in tqdm(to_counter.items(), desc="Checking nodes")
    if count == 1 and node not in from_nodes
]

print(f"Found {len(special_nodes)} special nodes.")

with open(output_file, 'w', encoding='utf-8') as f:
    f.write("to_node\ttimestamp\tblockNumber\tfrom_address\tgasUsed\tgasLimit\n")
    f.writelines(
        f"{node}\t{time}\t{block}\t{from_addr}\t{gas_used}\t{gas_limit}\n"
        for node, time, block, from_addr, gas_used, gas_limit in special_nodes
    )

print(f"Done! Output saved to: {output_file}")
```

```python
import os
import pandas as pd
from tqdm import tqdm  

input_folder = r"D:\sai\address_mapped_final_2"  

output_file = r"D:\sai\revert_tx_timestamps_blocks.txt"

txt_files = [f for f in os.listdir(input_folder) if f.endswith(".txt")]

with open(output_file, 'w') as out_file:
    out_file.write("timestamp\tblockNumber\n")  
    for filename in tqdm(txt_files, desc="Processing files"):
        file_path = os.path.join(input_folder, filename)

        try:
            df = pd.read_csv(file_path, sep='\t', dtype=str)
        except Exception as e:
            print(f"Skipping {filename} due to error: {e}")
            continue

        if not {'timestamp', 'blockNumber', 'isError'}.issubset(df.columns):
            print(f"Skipping {filename}: missing required columns.")
            continue

        revert_df = df[df['isError'] == '1']

        for _, row in revert_df.iterrows():
            out_file.write(f"{row['timestamp']}\t{row['blockNumber']}\n")

print(f"\n Revert transactions saved to: {output_file}")
```

```python
import os
import csv
from collections import Counter
from tqdm import tqdm

input_folder = r"D:\sai\address_mapped_final_2" 
output_file = r"D:\sai\calling_function_counts.txt"

function_counter = Counter()

files = [f for f in os.listdir(input_folder) if f.endswith(".txt")]

for filename in tqdm(files, desc="Processing files"):
    filepath = os.path.join(input_folder, filename)
    try:
        with open(filepath, 'r', encoding='utf-8') as f:
            reader = csv.DictReader(f, delimiter='\t')
            for row in reader:
                func = row.get("callingFunction", "").strip()
                if func:  
                    function_counter[func] += 1
    except Exception as e:
        print(f"Skipping {filename} due to error: {e}")

with open(output_file, 'w', encoding='utf-8') as f:
    f.write("callingFunction\tcount\n")
    for func, count in function_counter.most_common():
        f.write(f"{func}\t{count}\n")

print(f"\n Function call counts saved to: {output_file}")
```

```python
import os
import csv
from collections import defaultdict, Counter
from tqdm import tqdm

input_folder = r"D:\sai\address_mapped_final_2" 
output_file = r"D:\sai\most_common_calling_function_per_block.txt"
block_function_counter = defaultdict(Counter)

files = [f for f in os.listdir(input_folder) if f.endswith(".txt")]

for filename in tqdm(files, desc="Analyzing function counts"):
    filepath = os.path.join(input_folder, filename)
    try:
        with open(filepath, 'r', encoding='utf-8') as f:
            reader = csv.DictReader(f, delimiter='\t')
            for row in reader:
                block = row.get("blockNumber", "").strip()
                func = row.get("callingFunction", "").strip()
                if block and func:
                    block_function_counter[block][func] += 1
    except Exception as e:
        print(f"Error in {filename}: {e}")

results = []
for block, counter in block_function_counter.items():
    most_common_func, count = counter.most_common(1)[0]
    results.append((block, most_common_func, count))

with open(output_file, 'w', encoding='utf-8') as f:
    f.write("blockNumber\tcallingFunction\tcount\n")
    for block, func, count in sorted(results, key=lambda x: int(x[0])): 
        f.write(f"{block}\t{func}\t{count}\n")

print(f"\n Most common calling function per block saved to:\n{output_file}")
```

```python
import os
import csv
from collections import defaultdict
from tqdm import tqdm
import pandas as pd

input_folder = r"D:\sai\address_mapped_final_2"  
output_file = r"D:\sai\transactions_per_block.txt"

block_transaction_count = defaultdict(int)

all_files = [f for f in os.listdir(input_folder) if f.endswith(".txt")]

for file in tqdm(all_files, desc="Processing files"):
    file_path = os.path.join(input_folder, file)
    with open(file_path, 'r', encoding='utf-8') as f:
        reader = csv.reader(f, delimiter='\t')
        header = next(reader)

        if "blockNumber" not in header:
            continue  

        idx_block = header.index("blockNumber")

        for row in reader:
            if len(row) > idx_block:
                block_number = row[idx_block]
                block_transaction_count[block_number] += 1


with open(output_file, 'w', encoding='utf-8') as f:
    f.write("blockNumber\tcount\n")
    for block, count in sorted(block_transaction_count.items(), key=lambda x: int(x[0])):
        f.write(f"{block}\t{count}\n")

print(f"Saved transaction count per block to: {output_file}")
```
